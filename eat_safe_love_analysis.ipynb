### Notebook Setup

```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
import json
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# Assign the uk_food database to a variable name
db = mongo['uk_food']
```

```python
# Review the collections in database
print(db.list_collection_names())
```

```python
# Assign the collection to a variable
establishments = db['establishments']
establishments.count_documents({})
```

## Part 3: Exploratory Analysis

#### Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': 20}

results = establishments.find(query)

# Use count_documents to display the number of documents in the result
print(f"There are {establishments.count_documents(query)} establishments with a hygiene score of 20.")

# Display the first document in the results using pprint
pprint(results[0])
```

```python
# Put results into a list
data = list(results)
```

```python
# Add the results to a Pandas DataFrame
poor_hygiene_df = pd.DataFrame(data)

#Display the first 10 rows
poor_hygiene_df.head(10)
```
